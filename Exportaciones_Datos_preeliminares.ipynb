```
import pandas as pd

archivo_excel = 'Consulta_20260320-100018874.xlsx'
df = pd.read_excel(archivo_excel, sheet_name=0, skiprows=9)
df = df.iloc[:, [0, 1]]
df.columns = ['Fecha', 'Valor']
df
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')
df['Valor'] = df['Valor'].astype(str).str.replace(',', '').str.strip()
df['Valor'] = pd.to_numeric(df['Valor'], errors='coerce')
df['Año'] = df['Fecha'].dt.year
df_limpio = df.dropna(subset=['Año', 'Valor'])
sumas_reales = df_limpio.groupby('Año')['Valor'].sum().reset_index()
pd.options.display.float_format = '{:,.2f}'.format
print("Sumatorias de Importaciones Totales por Año (Calculado con precisión):")
display(sumas_reales)
```

```
#1993-2000
p1 = df[(df['Año'] >= 1993) & (df['Año'] <= 2000)]
sumatoria_p1 = p1.groupby('Año')['Valor'].sum().reset_index()
sumatoria_p1.columns = ['Año', 'Suma Importaciones (1993-2000)']

print("SUMATORIAS PERIODO 1993-2000")
display(sumatoria_p1)
```

```
 #2011-2020
p3 = df[(df['Año'] >= 2011) & (df['Año'] <= 2020)]
sumatoria_p3 = p3.groupby('Año')['Valor'].sum().reset_index()
sumatoria_p3.columns = ['Año', 'Suma Importaciones (2011-2020)']

print(" SUMATORIAS PERIODO 2011-2020")
display(sumatoria_p3)
```

```
# Filtramos y sumamos para 2021-2026
p4 = df[(df['Año'] >= 2021) & (df['Año'] <= 2026)]
sumatoria_p4 = p4.groupby('Año')['Valor'].sum().reset_index()
sumatoria_p4.columns = ['Año', 'Suma Importaciones (2021-2026)']

print("SUMATORIAS PERIODO 2021-2026")
display(sumatoria_p4)
```

```
import pandas as pd
import numpy as np

# 1. Identificar columna de fecha automáticamente (Evita errores de nombres)
posibles_nombres = ['Fecha', 'fecha', 'Date', 'date', 'FECHA']
col_fecha = next((c for c in posibles_nombres if c in df.columns), None)

if col_fecha:
    # 2. Convertir a fecha real (Rúbrica: Proceso bien justificado)
    # 'dayfirst=True' ayuda con el formato común en México (DD/MM/AAAA)
    df[col_fecha] = pd.to_datetime(df[col_fecha], dayfirst=True, errors='coerce')

    # 3. Eliminar filas con fechas que no se pudieron arreglar (Tratamiento de NA)
    df = df.dropna(subset=[col_fecha])

    # 4. Ordenar y establecer índice (Requisito para Series de Tiempo) [cite: 15, 32]
    df = df.sort_values(col_fecha).set_index(col_fecha)

    # 5. Rellenar huecos y promediar duplicados (Resampling) [cite: 76]
    # 'D' significa frecuencia diaria. Esto crea una serie continua para el modelo.
    df = df.resample('D').mean().interpolate(method='linear')

    print(f"¡Limpieza exitosa! Columna detectada: '{col_fecha}'")
    print(f"Rango de datos: desde {df.index.min()} hasta {df.index.max()}")
else:
    print(" ERROR: No encontré una columna llamada 'Fecha'. Revisa tu Excel.")

# Mostrar los datos listos para el modelo
df.head()
```

```
# --- PREDICCIÓN DEL POLLO --- (Fixed) ---

# Reload and clean the data correctly within this context for Prophet
# This assumes 'base de datos.xlsx' is the correct file and has the structure
# that needs skiprows=9 to get to the actual data.
# The 'uploaded' and 'file_name' variables are available from cell 'OeNWKkg_Moc7'.

import io

df_temp = pd.read_excel(io.BytesIO(uploaded[file_name]), skiprows=9)

# Select the first two columns (Fecha and Valor) and rename them
df_temp = df_temp.iloc[:, [0, 1]]
df_temp.columns = ['Fecha', 'Valor']

# Clean 'Fecha' column and set as index
df_temp['Fecha'] = pd.to_datetime(df_temp['Fecha'], dayfirst=True, errors='coerce')
df_temp = df_temp.dropna(subset=['Fecha']).sort_values('Fecha').set_index('Fecha')

# Resample to daily frequency and interpolate missing values
df_clean_prophet = df_temp.resample('D').mean().interpolate(method='linear')

# Prepare data for Prophet (requires 'ds' for dates and 'y' for values)
df_p = df_clean_prophet.reset_index().rename(columns={'Fecha': 'ds', 'Valor': 'y'})

# Drop rows where 'y' is NaN after cleaning, as Prophet cannot handle them
df_p = df_p.dropna(subset=['y'])

model_p = Prophet(yearly_seasonality=True)
model_p.fit(df_p)

# Horizonte de 6 meses (Requisito del Proyecto Prototípico)
future_p = model_p.make_future_dataframe(periods=6, freq='M')
forecast_p = model_p.predict(future_p)

# Visualización (Rúbrica: Visualizaciones claras)
fig = model_p.plot(forecast_p)
plt.title("Pronóstico de Precios: Pollo")
plt.show()
```

```
def estrategia_mercado_justo(actual, futuro, costo_guarda=1.5):
    """
    Determina si conviene vender basado en la predicción.
    """
    diferencia = futuro - actual

    print(f"--- Análisis de Comercialización Justa ---")
    if diferencia > costo_guarda:
        return "RECOMENDACIÓN: ALMACENAR. El precio subirá más que el costo de guarda."
    else:
        return "RECOMENDACIÓN: VENDER AHORA. No se espera un alza suficiente."

# Ejemplo con el último dato real y la última predicción
p_actual = df_pollo['y'].iloc[-1]
p_futuro = forecast_p['yhat'].iloc[-1]

print(estrategia_mercado_justo(p_actual, p_futuro))
```

```
# Carga de archivo local
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
# Load the Excel file, skipping the first 9 rows as determined in previous successful loads
df = pd.read_excel(io.BytesIO(uploaded[file_name]), skiprows=9)

# Select the first two columns and rename them to 'Fecha' and 'Valor'
df = df.iloc[:, [0, 1]]
df.columns = ['Fecha', 'Valor']

# Limpieza técnica
df['Fecha'] = pd.to_datetime(df['Fecha'], dayfirst=True, errors='coerce')
df = df.dropna(subset=['Fecha']).sort_values('Fecha').set_index('Fecha')

# Resampling y manejo de Nulos (Justificado para mantener la tendencia)
df_clean = df.resample('D').mean().interpolate(method='linear')
print("✅ Serie temporal lista para el modelado.")
```

```
# Preparación para Prophet
df_p = df_clean[['Valor']].reset_index().rename(columns={'Fecha':'ds', 'Valor':'y'})

modelo = Prophet(yearly_seasonality=True)
modelo.fit(df_p)

# Predicción a 6 meses (Horizonte solicitado)
futuro = modelo.make_future_dataframe(periods=180)
forecast = modelo.predict(futuro)

# Gráfico de predicción profesional
fig = modelo.plot(forecast)
plt.title("Pronóstico de Precios: Próximos 6 Meses")
```
